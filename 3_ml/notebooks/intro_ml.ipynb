## 2 · ¿Cuántos grupos? Validación con métricas

Un *clustering* no tiene "respuesta correcta" con la que compararse. Aun así,
queremos **cuantificar** lo bueno que es. Dos medidas habituales:

- **WCSS** (inercia): suma de distancias² de cada punto al centro de su cluster.
  *Cuanto más pequeña, mejor.*
- **Silhouette**: para cada punto, mide si está más cerca de su grupo que de los
  demás. *Cuanto más grande (máx. 1), mejor.*

---
### 🎯 Lo que os lleváis

- **No supervisado**: agrupar (*k-means*, validado con **codo**/**silhouette**) y
  reducir variables (**PCA**) — sin decirle al ordenador la respuesta.
- **Supervisado**: **regresión lineal** (simple e interpretable) y **árboles** (y sus
  *ensembles*: random forest, boosting).
- **La regla de oro**: evaluad **siempre en test**, nunca con los datos de
  entrenamiento (o el *overfitting* os engañará).
- **Ingeniería de variables** (one-hot, escalado…) para que el modelo entienda mejor.
- Y todo esto **dirigiendo a un agente** y **comprobando** lo que devuelve.
