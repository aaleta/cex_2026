```python
# === Datos ===  (ejecútala tal cual)
DATA_URL = "https://raw.githubusercontent.com/aaleta/cex_2026/main/0_introduccion/data/"  # cámbialo si usas otro repo: "https://raw.githubusercontent.com/.../data/"
ARCHIVO  = "unizar_grados.csv.gz"

import os, urllib.request
def datos(nombre=ARCHIVO):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        try:
            from google.colab import files            # subir desde tu ordenador
            print(f"Sube '{nombre}' desde tu ordenador…"); files.upload()
        except Exception:
            raise FileNotFoundError(f"No encuentro '{nombre}'. Súbelo a Colab.")
    return nombre

datos()
print("✅ Datos listos:", ARCHIVO)
```

### ¿Qué significa cada columna?

| columna | qué es |
|---|---|
| `year` | curso académico (2013 = curso 2013-14) |
| `degree` | grado (p. ej. *Medicina*, *Economía*) |
| `faculty` | centro / facultad |

| `subject_type` | tipo (*Obligatoria*, *Optativa*, *Formación básica*…) |
| `students` | nº de estudiantes **matriculados** |
| `presented` | nº que se **presentaron** al examen |

| `success_rate` | % que aprueba **de los que se presentan** (`passed/presented`) |

| `eval_rate` | % que se presenta (`presented/students`) |
| `avg_attempts` | nº medio de convocatorias usadas |
