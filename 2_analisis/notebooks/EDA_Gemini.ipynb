### **2. Análisis de Características Categóricas**

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

plt.figure(figsize=(18, 12))
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 2, i + 1)
    sns.countplot(data=df, x=feature, palette='viridis')
    plt.title(f'Distribución de {feature}')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### **3. Análisis de Características Numéricas**

```python
numerical_features = ['math score', 'reading score', 'writing score']

plt.figure(figsize=(15, 5))
for i, feature in enumerate(numerical_features):
    plt.subplot(1, 3, i + 1)
    sns.histplot(data=df, x=feature, kde=True, bins=20, palette='magma')
    plt.title(f'Distribución de {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
for i, feature in enumerate(numerical_features):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(data=df, y=feature, palette='coolwarm')
    plt.title(f'Boxplot de {feature}')
plt.tight_layout()
plt.show()
```

### **4. Relación entre Variables Numéricas y Categóricas**

```python
plt.figure(figsize=(18, 15))
for i, cat_feature in enumerate(categorical_features):
    for j, num_feature in enumerate(numerical_features):
        plt.subplot(len(categorical_features), len(numerical_features), i * len(numerical_features) + j + 1)
        sns.boxplot(data=df, x=cat_feature, y=num_feature, palette='Paired')
        plt.title(f'{num_feature} por {cat_feature}')
        plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## **Visualizaciones**

#### **5.1. Matriz de Correlación de Puntuaciones**

Esta visualización nos ayudará a entender cómo se relacionan las puntuaciones de las diferentes asignaturas entre sí.

```python
correlation_matrix = df[numerical_features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación entre Puntuaciones')
plt.show()
```

#### **5.2. Puntuaciones por Nivel Educativo de los Padres y Género (Violin Plots)**

Los gráficos de violín son excelentes para mostrar la distribución de las puntuaciones para diferentes grupos categóricos, revelando no solo la mediana y los cuartiles, sino también la densidad de los datos.

```python
plt.figure(figsize=(18, 10))

plt.subplot(1, 3, 1)
sns.violinplot(data=df, x='parental level of education', y='math score', hue='gender', palette='muted')
plt.title('Puntuación de Matemáticas por Educación Parental y Género')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Género')

plt.subplot(1, 3, 2)
sns.violinplot(data=df, x='parental level of education', y='reading score', hue='gender', palette='muted')
plt.title('Puntuación de Lectura por Educación Parental y Género')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Género')

plt.subplot(1, 3, 3)
sns.violinplot(data=df, x='parental level of education', y='writing score', hue='gender', palette='muted')
plt.title('Puntuación de Escritura por Educación Parental y Género')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Género')

plt.tight_layout()
plt.show()
```
