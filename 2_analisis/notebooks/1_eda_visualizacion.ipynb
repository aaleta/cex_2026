## 5 · ¿Cómo voy respecto a la clase? Posición relativa (Z-score)

Un 70 en matemáticas… ¿es bueno? Depende del resto. El **Z-score** (o
*standard score*) contesta a *"¿a cuántas desviaciones típicas de la media
estoy?"*:

$$Z = \frac{x - \mu}{\sigma}$$

donde $\mu$ es la media y $\sigma$ la desviación típica. Un $Z=0$ es justo la
media; $Z=+1$, una desviación por encima.
