# Merge de dos tablas distintas

Vamos a juntar dos tablas distintas:

- Una tabla con ofertas de trabajo
- Otra tabla con información sobre las empresas

Además en concreto buscamos responder a la pregunta: ¿Cuántos trabajos hay en Research & Development?

```python
# === Datos ===
DATA_URL = "https://raw.githubusercontent.com/aaleta/cex_2026/main/1_obtencion/data/"

import os, urllib.request
def datos(nombre):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"Falta '{nombre}': súbelo a Colab o define DATA_URL")
    return nombre

for _f in ['jobs_data.csv', 'companies_data.json']:
    datos(_f)
print('Datos listos:', ['jobs_data.csv', 'companies_data.json'])
```
