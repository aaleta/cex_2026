# M3 · Las variables y el objetivo

Convertimos la serie temporal en una tabla lista para *machine learning*. Aquí se decide casi todo: **qué** predecir, **con qué**, y sin hacernos trampas.

```python
# === Datos ===
DATA_URL = "https://raw.githubusercontent.com/aaleta/bizi_curso/main/data/" 

import os, urllib.request
def datos(nombre):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"Falta '{nombre}': súbelo a Colab o define DATA_URL")
    return nombre

for _f in ['bizi_15min_2weeks.csv', 'bizi_stations.csv']:
    datos(_f)
print('Datos listos:', ['bizi_15min_2weeks.csv', 'bizi_stations.csv'])
```
