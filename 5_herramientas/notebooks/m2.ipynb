Un buen gráfico enseña más que una tabla. Vamos a ver el *pulso* del sistema Bizi: cuándo y dónde hay bicis, y por qué.

```python
# === Datos ===
DATA_URL = "https://raw.githubusercontent.com/aaleta/cex_2026/main/5_herramientas/data/"

import os, urllib.request
def datos(nombre):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"Falta '{nombre}': súbelo a Colab o define DATA_URL")
    return nombre

for _f in ['bizi_15min.parquet', 'bizi_stations.csv']:
    datos(_f)
print('Datos listos:', ['bizi_15min.parquet', 'bizi_stations.csv'])
```
