# Apéndice · El clima (bonus)

La mejor variable suele estar **fuera** del fichero. El clima debería mover la demanda… veamos si de verdad ayuda al modelo.

```python
# === Datos ===
DATA_URL = "https://raw.githubusercontent.com/aaleta/bizi_curso/main/data/"

import os, urllib.request
def datos(nombre):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"Falta '{nombre}': súbelo a Colab o define DATA_URL")
    return nombre

for _f in ['bizi_model.parquet', 'bizi_15min.parquet']:
    datos(_f)
print('Datos listos:', ['bizi_model.parquet', 'bizi_15min.parquet'])
```

## 1 · Descargar el tiempo de Zaragoza (Open-Meteo, gratis, sin clave)

```python
import pandas as pd, numpy as np, json, urllib.request
url = ("https://archive-api.open-meteo.com/v1/archive?latitude=41.65&longitude=-0.88"
       "&start_date=2026-03-17&end_date=2026-07-11"
       "&hourly=temperature_2m,precipitation,wind_speed_10m&timezone=UTC")
h = json.load(urllib.request.urlopen(url))["hourly"]
clima = pd.DataFrame({"hour_utc": pd.to_datetime(h["time"]), "temp": h["temperature_2m"],
                      "precip": h["precipitation"], "wind": h["wind_speed_10m"]})
print(clima.shape, "| temp entre", clima.temp.min(), "y", clima.temp.max(), "°C")
clima.head()
```
