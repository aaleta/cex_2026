# M8 · Despliegue (debate, sin código)

Ya tienes un modelo entrenado y evaluado. Pero un modelo en un notebook **no sirve a nadie**. Este bloque es de **conceptos y debate**: no hay que programar.

## Bloque A · La parte técnica
1. **¿Dónde corre?** tu equipo · servidor propio (*on-premises*) · nube (AWS/Google/Azure) · *serverless*.
2. **¿Batch u online?** puntuar todo cada pocos min y guardar, vs. una **API** que responde al instante.
3. **¿Dónde viven los datos?** el *scraper* escribe en una **base de datos**; el modelo lee de ahí y escribe sus predicciones; el panel/aviso las lee.
4. **¿Cómo se controla?** horario (cron) · control de versiones · *rollback* · el **umbral** como ajuste.

## Bloque B · Mantenerlo vivo
5. **Model drift:** un modelo entrenado en primavera empeora en verano → **reentrenar**.
6. **Monitorización:** vigilar entradas (¿scraper vivo?, ¿sensores obsoletos? ¡M1!) y salidas (¿sigue acertando?).
7. **Skew online/offline:** calcular las variables en vivo **igual** que al entrenar (¡otra vez la zona horaria de M1!).
