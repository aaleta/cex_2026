# M4 · Modelos baseline

¿Cuál es el resultado **mínimo** que hay que superar? Eso es un *baseline*. Un buen baseline lleva sentido común dentro, y a veces cuesta batirlo.

```python
# === Datos ===
DATA_URL = "https://raw.githubusercontent.com/aaleta/cex_2026/main/5_herramientas/data/"

import os, urllib.request
def datos(nombre):
    if not os.path.exists(nombre) and DATA_URL:
        urllib.request.urlretrieve(DATA_URL + nombre, nombre)
    if not os.path.exists(nombre):
        raise FileNotFoundError(f"Falta '{nombre}': súbelo a Colab o define DATA_URL")
    return nombre

for _f in ['bizi_model.parquet']:
    datos(_f)
print('Datos listos:', ['bizi_model.parquet'])
```

```python
import pandas as pd, numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
m = pd.read_parquet("bizi_model.parquet")
tr, te = m[m.split=="train"], m[m.split=="test"]
y = te.empty_in_30min
def evalua(nombre, pred):
    print(f"{nombre:34} acc={accuracy_score(y,pred):.3f} "
          f"prec={precision_score(y,pred,zero_division=0):.3f} "
          f"rec={recall_score(y,pred,zero_division=0):.3f} f1={f1_score(y,pred,zero_division=0):.3f}")
print("base:", f"{y.mean():.1%} de las filas se vacían")
```
